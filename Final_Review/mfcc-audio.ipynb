```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate and plot MFCCs
def calculate_and_plot_mfcc(audio_file, sr=22050, n_mfcc=13):
    # Load the audio file
    y, sr = librosa.load(audio_file, sr=sr)
    
    # Calculate MFCCs
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    
    # Plot the MFCCs
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfccs, x_axis='time', sr=sr, cmap='viridis')
    plt.colorbar(format='%+2.0f dB')
    plt.title('MFCCs')
    plt.tight_layout()
    plt.show()
    
    return mfccs
```

```python
audio_file_1 = "/kaggle/input/ben-acc/bengali_accent_hin_2_2.wav"  # Replace with the path to your first audio file
audio_file_2 = "/kaggle/input/gt-data-bengali/GT_bengali_accent_hin_2_2.wav"  # Replace with the path to your second audio file

# Calculate and plot MFCCs for the first audio
print("MFCCs for Audio 1:")
mfccs_1 = calculate_and_plot_mfcc(audio_file_1)

# Calculate and plot MFCCs for the second audio
print("MFCCs for Audio 2:")
mfccs_2 = calculate_and_plot_mfcc(audio_file_2)

# Optional: Compare MFCCs
print("MFCC shape for Audio 1:", mfccs_1.shape)
print("MFCC shape for Audio 2:", mfccs_2.shape)
```
