```python
import nemo.collections.asr as nemo_asr
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
model = nemo_asr.models.EncDecCTCModel.restore_from(restore_path='/home/arunb/usg_lab/bhashika/Models/ai4b_indicConformer_hi.nemo')
model.freeze()
model = model.to(device)
```

```python
import pandas as pd
import requests
import os
import ffmpeg
import torch  # Assuming your model uses PyTorch for GPU handling

# Check if a GPU is available
if torch.cuda.is_available():
    print("GPU is available. Model will use GPU for transcription.")
else:
    print("Warning: GPU is not available. This will run on CPU, which might be slower.")

# Load or create the transcription CSV file
csv_file_path = 'Bihar_final_updated.csv'  # Replace with the actual CSV path
csv_file_path1 = 'Bihar_final_updated.csv'

# Load the original CSV
df = pd.read_csv(csv_file_path)
print(df.shape)

# Check if the transcribed file exists
if os.path.exists(csv_file_path1):
    df1 = pd.read_csv(csv_file_path1)
else:
    # If it doesn't exist, create it by copying the original and adding necessary columns
    df1 = df.copy()
    df1['downsampled_path'] = None
    df1['transcription'] = None
    df1.to_csv(csv_file_path1, index=False)

# Create output folders if they don't exist
output_folder = 'bihar_files'
output_folder1 = 'downsampled_files'
os.makedirs(output_folder, exist_ok=True)
os.makedirs(output_folder1, exist_ok=True)

# Function to download file
def download_file(url, save_path):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(save_path, 'wb') as f:
            f.write(response.content)
        return True
    return False

# Loop through each row in the CSV
for index, row in df.iterrows():
    if index <= 72582:
        continue
    
    file_url = row['file_url']
    file_name = row['file_name']
    file_path = os.path.join(output_folder, file_name)
    
    # Check if the .wav file already exists, if not, download it
    if not os.path.exists(file_path):
        download_success = download_file(file_url, file_path)
        if not download_success:
            print(f"Failed to download {file_url}")
            continue
    else:
        print(f"File {file_name} already exists. Skipping download.")
    
    # Downsample and convert to single channel
    downsampled_path = os.path.join(output_folder1, f"audio_{index}_infer_ready.wav")
    ffmpeg.input(file_path).output(downsampled_path, ac=1, ar=16000).run(overwrite_output=True)
    df1.at[index, 'downsampled_path'] = downsampled_path

    # Transcribe the audio on GPU
    try:
        model.cur_decoder = "ctc"
        transcription = model.transcribe([downsampled_path], batch_size=1, logprobs=False, language_id='hi')[0]
        df1.at[index, 'transcription'] = transcription
    except Exception as e:
        print(f"Error transcribing {file_name}: {e}")
        df1.at[index, 'transcription'] = None

    # Save progress incrementally
    df1.to_csv(csv_file_path1, index=False)
    print(f"Processed file {file_name} and saved transcription.")

print("All files processed.")
```
