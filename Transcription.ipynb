```python
import requests
import os
import pandas as pd  

def download_wav_file(url, save_folder, file_name):
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

    response = requests.get(url)

    if response.status_code == 200:
        file_path = os.path.join(save_folder, f"{file_name}")

        # Save the content as a .wav file
        with open(file_path, "wb") as file:
            file.write(response.content)

        print(f"File saved at: {file_path}")
        return file_path  # Return the file path for further processing
    else:
        print(f"Failed to download file. Status code: {response.status_code}")
        return None  # Return None if the download failed

# Example usage
save_folder = os.path.join("/kaggle/working/", "Bihar_Dataset_wavFiles")
os.makedirs(save_folder, exist_ok=True)
model.cur_decoder = "ctc"

# Create a new column "transcript" in the DataFrame if it doesn't exist
if 'transcript' not in df.columns:
    df['transcript'] = None

batch_size = 30  # Set the batch size to append after every 30 rows
processed_rows = 0  # Initialize a counter for processed rows
accumulated_rows = []  # List to accumulate processed rows

# Define the CSV file path
csv_file_path = '/kaggle/working/transcribed_data_15_10_1819.csv'

# Process rows starting from index 4928 till the end of the DataFrame
for i in range(36796, len(df)):
    url = df["file_url"][i]
    file_name = df['file_name'][i]

    # Check if the URL is valid, skip the row if not
    if pd.isna(url) or not url:
        print(f"Skipping row {i} due to missing URL.")
        continue

    # Download the wav file
    wav_path = download_wav_file(url, save_folder, file_name)
    
    # If the file was successfully downloaded
    if wav_path:
        try:
            # Perform transcription using the model
            ctc_text = model.transcribe([wav_path], batch_size=1, logprobs=False, language_id='hi')[0]
            
            # Store the transcription in the "transcript" column
            df.at[i, 'transcript'] = ctc_text[0]
            processed_rows += 1  # Increment the processed rows counter

            # Print the index, file name (ID), and transcript to the console
            print(f"Index: {i}, File Name: {file_name}, Transcript: {ctc_text[0]}")

            # Accumulate the processed row
            accumulated_rows.append(df.iloc[i])

            # Append accumulated rows after every 30 processed rows
            if processed_rows >= batch_size:
                pd.DataFrame(accumulated_rows).to_csv(csv_file_path, mode='a', index=False, header=not os.path.exists(csv_file_path))
                accumulated_rows = []  # Clear the list after appending
                processed_rows = 0  # Reset the counter after appending

        except Exception as e:
            print(f"Error in transcription for row {i}: {e}")
    else:
        print(f"File not downloaded for row {i}, skipping transcription.")

# After the loop, ensure any remaining rows are written to the file
if accumulated_rows:
    pd.DataFrame(accumulated_rows).to_csv(csv_file_path, mode='a', index=False, header=not os.path.exists(csv_file_path))
```
